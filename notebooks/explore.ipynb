```python
import sqlite3
import pandas as pd
from pathlib import Path
pd.set_option("display.max_colwidth", None) 

DB_PATH = Path("reviews.db")
conn = sqlite3.connect(DB_PATH)
print("✅ Подключено к", DB_PATH)
```

```python
# Таблицы в БД
pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
```

```python
# Схема таблицы reviews
pd.read_sql("PRAGMA table_info(reviews)", conn)
```

```python
pd.read_sql("PRAGMA table_info(categories)", conn)
```

```python
pd.read_sql("SELECT * FROM categories ORDER BY group_type, count DESC", conn)
```

```python
# Общая статистика
pd.read_sql("""
    SELECT
        COUNT(*)                        AS total,
        MIN(published_date)             AS oldest,
        MAX(published_date)             AS newest,
        ROUND(AVG(rating), 2)           AS avg_rating,
        SUM(reply IS NOT NULL)          AS with_reply
    FROM reviews
""", conn)
```

```python
# Распределение по рейтингу
pd.read_sql("""
    SELECT rating, COUNT(*) AS count
    FROM reviews
    GROUP BY rating
    ORDER BY rating DESC
""", conn)
```

```python
# Все отзывы
df = pd.read_sql("SELECT * FROM reviews ORDER BY published_date DESC", conn)
df
```

```python
# Негативные отзывы (рейтинг ≤ 3)
pd.read_sql("""
    SELECT id, rating, title, text, published_date
    FROM reviews
    WHERE rating <= 3
    ORDER BY published_date DESC
""", conn)
```

```python
# Отзывы с ответом компании
pd.read_sql("""
    SELECT id, rating, title, reply, reply_date
    FROM reviews
    WHERE reply IS NOT NULL
    ORDER BY reply_date DESC
""", conn)
```

```python
# Закрыть соединение когда закончил
conn.close()
print("🔒 Соединение закрыто")
```
