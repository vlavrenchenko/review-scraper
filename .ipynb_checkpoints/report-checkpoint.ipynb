```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
from pathlib import Path

pd.set_option('display.max_colwidth', None)
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 10

DB_PATH = Path('reviews.db')
conn = sqlite3.connect(DB_PATH)
print('✅ Подключено к', DB_PATH)
```

```python
# Загружаем отзывы и категории
reviews = pd.read_sql(
    'SELECT company, rating, reply FROM reviews WHERE rating IS NOT NULL',
    conn
)

# Берём только последний прогон анализа для каждой компании
latest_runs = pd.read_sql("""
    SELECT company, MAX(analyzed_at) AS last_run
    FROM categories
    GROUP BY company
""", conn)

categories = pd.read_sql("""
    SELECT c.company, c.group_type, c.name, c.count, c.analyzed_at
    FROM categories c
    JOIN (
        SELECT company, MAX(analyzed_at) AS last_run
        FROM categories GROUP BY company
    ) lr ON c.company = lr.company AND c.analyzed_at = lr.last_run
""", conn)

# Порядок компаний
companies = reviews['company'].unique().tolist()
print(f'Компании: {companies}')
print(f'Отзывов: {len(reviews)}, категорий: {len(categories)}')
```

## 1. Распределение оценок по компаниям

```python
rating_dist = (
    reviews.groupby(['company', 'rating'])
    .size()
    .unstack(fill_value=0)
    .reindex(columns=[1, 2, 3, 4, 5], fill_value=0)
)

# Нормируем в проценты
rating_pct = rating_dist.div(rating_dist.sum(axis=1), axis=0) * 100

fig, ax = plt.subplots(figsize=(10, 5))

x = np.arange(len(rating_pct))
width = 0.15
colors = ['#e74c3c', '#e67e22', '#f1c40f', '#2ecc71', '#27ae60']

for i, rating in enumerate([1, 2, 3, 4, 5]):
    ax.bar(x + i * width, rating_pct[rating], width, label=f'★ {rating}', color=colors[i])

ax.set_xticks(x + width * 2)
ax.set_xticklabels(rating_pct.index, rotation=15, ha='right')
ax.set_ylabel('Доля отзывов, %')
ax.set_title('Распределение оценок по компаниям')
ax.legend(title='Оценка', bbox_to_anchor=(1.01, 1), loc='upper left')
ax.yaxis.set_major_formatter(mticker.FormatStrFormatter('%.0f%%'))
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

rating_pct.round(1)
```

## 2. Топ-5 категорий негативных отзывов по компаниям

```python
neg_cats = (
    categories[categories['group_type'] == 'negative']
    .sort_values(['company', 'count'], ascending=[True, False])
    .groupby('company')
    .head(5)
)

company_list = neg_cats['company'].unique()
n = len(company_list)
fig, axes = plt.subplots(1, n, figsize=(5 * n, 5), sharey=False)
if n == 1:
    axes = [axes]

for ax, company in zip(axes, company_list):
    data = neg_cats[neg_cats['company'] == company].head(5)
    ax.barh(data['name'][::-1], data['count'][::-1], color='#e74c3c', alpha=0.8)
    ax.set_title(company, fontweight='bold')
    ax.set_xlabel('Упоминаний')
    ax.xaxis.set_major_locator(mticker.MaxNLocator(integer=True))
    ax.grid(axis='x', alpha=0.3)

fig.suptitle('Топ-5 категорий негативных отзывов', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

neg_cats[['company', 'name', 'count']]
```

## 3. Доля негативных отзывов с ответом поддержки

```python
negative = reviews[reviews['rating'] <= 3].copy()

reply_stats = (
    negative.groupby('company')
    .agg(
        total=('rating', 'count'),
        with_reply=('reply', lambda x: x.notna().sum())
    )
    .assign(reply_rate=lambda df: df['with_reply'] / df['total'] * 100)
    .sort_values('reply_rate', ascending=True)
)

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(reply_stats.index, reply_stats['reply_rate'], color='#3498db', alpha=0.8)

for bar, (_, row) in zip(bars, reply_stats.iterrows()):
    ax.text(
        bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
        f"{row['reply_rate']:.1f}%  ({int(row['with_reply'])}/{int(row['total'])})",
        va='center', fontsize=9
    )

ax.set_xlabel('Доля негативных отзывов с ответом, %')
ax.set_title('Доля негативных отзывов с ответом поддержки')
ax.set_xlim(0, 110)
ax.xaxis.set_major_formatter(mticker.FormatStrFormatter('%.0f%%'))
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

reply_stats
```
